```python
from google.colab import files
from PIL import Image
import os
from IPython.display import Image as IPImage, display

print(os.listdir("frames"))

folder = "frames"
frames = []

# Get sorted image list
image_files = sorted([
    f for f in os.listdir(folder)
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
])

print("Images found:", image_files)

# Open first image
first_image = Image.open(os.path.join(folder, image_files[0])).convert("RGB")
size = first_image.size
frames.append(first_image)

# Process remaining images
for filename in image_files[1:]:
    img = Image.open(os.path.join(folder, filename)).convert("RGB")
    img = img.resize(size)
    frames.append(img)

# SAVE GIF
frames[0].save(
    "output.gif",
    format="GIF",
    save_all=True,
    append_images=frames[1:],
    duration=200,
    loop=0
)

print("GIF created successfully")

display(IPImage("output.gif"))
files.download("output.gif")
```
